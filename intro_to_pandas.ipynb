<h3>Series</h3>
<ul><li>Tek boyutlu dizilerle uğraşmak için kullanılır.</li> <li>NumPy'dan farklı olarak dizideki değerler birbirinden farklı türlerde olabilir.</li>
<li>Değerlere indeksler yoluyla erişilir</li></ul>


```python
seri = pd.Series([1,2,3])               # sayısal indeksler
print(seri)
```

```python
sehirler = {"Bursa":16, "Yalova":77}    # metinsel indeksler
sehirler_seri = pd.Series(sehirler, name="Sehirler")
print(sehirler["Bursa"])
sehirler_seri
```

```python
satislar_df_eksiksiz = satislar_df.dropna(inplace=False)  # icerisinde eksik deger olan satırları sil
satislar_df_eksiksiz
```

```python
values = {"Birim_Fiyat":10, "Adet":0}
satislar_df_doldur = satislar_df.fillna(values)  # icerisinde eksik deger olan sütunları doldur
print(satislar_df_doldur)
```

<h3> Aritmetik İşlemler</h3>

```python
satislar_df.Birim_Fiyat += 100    # Birim fiyatı 100 TL arttır
satislar_df["Toplam_Satış"] = satislar_df.Birim_Fiyat * satislar_df.Adet  # Toplam Satışı Hesapla
satislar_df
```

<h3>Filtreleme</h3>

```python
filtre = satislar_df.Bolge == "Ege"    # Ege bölgesindeki satışlar
satislar_df[filtre]
```

```python
armut_filtre = (satislar_df.Urun == "Armut") & (satislar_df.Adet > 10)   # 10 dan fazla satılan armutlar
satislar_df[armut_filtre]
```

<h3>Sıralama</h3>

```python
satislar_df.sort_values(by="Adet", ascending=False)  # Adet sütununa göre sırala. Buyukten kucuge sırala
```

<h3>Sütundaki tüm değerlere bir fonksiyon uygulama</h3>

```python
def ilk_harf(metin):
    return metin[0]

satislar_df["İlk Harfler"] = satislar_df.Bolge.apply(ilk_harf)
satislar_df
```

<h3>Sütun Silme</h3>

```python
satislar_df.drop(columns=["Urun", "Adet"])   # Urun ve Adet sütunlarını sil
```
